# Part B
Add a column vector of 1’s to the features, then use the linear regression formula discussed in Lecture 3 to obtain a 2-dimensional weight vector. Plot both the linear regression line and the training data on the same graph. Also report the average error on the training set using Eq. (1).


```python
modifiedx1 = np.c_[ x1, np.ones(40) ]    

w= np.random.rand(40) 
```

```python
def hypothesis(x,w):

    return np.matmul(x,w)
```

```python
def costFn(x,w,y):
    return 1/2 *(hypothesis(x,w) - y)*(hypothesis(x,w) - y)
```

```python
def gradientDescent(X,y):
    return np.matmul(np.matmul(X.T*X))
```

```python
#Get weight vector
weights= gradientDescent(modifiedx1,y1)
```

```python
hypothesis(modifiedx1,weights)
```

```python
def averageError(w,x,y):
    return sum((hypothesis(w,x) - y)*((hypothesis(w,x) - y)))/x.shape[1]
```

```python
#Get average error
averageError(weights,modifiedx1.T,y1)
```
